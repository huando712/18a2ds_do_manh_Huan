đọc dữ liệu

```python
import pandas as pd
from pathlib import Path

# Đọc file CSV
csv_path = Path(r"D:\OneDrive\Máy tính\18a2ds_hn\18a2ds_do_manh_Huan\luyện tập\pandas\economic_social_on_tap.csv")  # đổi lại đúng đường dẫn của bạn
df = pd.read_csv(csv_path)

# Xem 5 dòng đầu
print(df.head())

# Kích thước dữ liệu
print("Shape:", df.shape)

# Thông tin cột & kiểu dữ liệu
print(df.info())
```

```python
# Số lượng giá trị thiếu theo từng cột
print(df.isna().sum())

# Tỷ lệ dữ liệu thiếu (%)
print((df.isna().mean() * 100).round(2))
```

```python
# Thống kê cơ bản
print(df.describe())
```

phân tích theo năm

```python
yearly_stats = df.groupby("year")[["population", "income", "cpi", "unemployment_rate"]].mean()
print(yearly_stats)
```

phân tích theo thành phố

```python
province_stats = df.groupby("province")[["population", "income", "unemployment_rate"]].mean()
print(province_stats)
```

```python
# 5 tỉnh có thu nhập trung bình cao nhất
top_income = province_stats.sort_values("income", ascending=False).head(5)
print(top_income)

# 5 tỉnh có tỷ lệ thất nghiệp cao nhất
top_unemployment = province_stats.sort_values("unemployment_rate", ascending=False).head(5)
print(top_unemployment)
```

```python
# Thu nhập trung bình theo năm
df.groupby("year")["income"].mean().plot(
    title="Thu nhập trung bình theo năm",
    xlabel="Year",
    ylabel="Income"
)
```

```python
import pandas as pd

# ===============================
# 1️⃣ ĐỌC DỮ LIỆU
# ===============================
df = pd.read_csv("D:/OneDrive/Máy tính/18a2ds_hn/18a2ds_do_manh_Huan/luyện tập/pandas/data_cleaned.csv")   # 🔴 đổi thành đường dẫn file CSV của bạn

print("===== 5 DÒNG ĐẦU =====")
print(df.head())

print("\n===== THÔNG TIN DỮ LIỆU =====")
print(df.info())

print("\n===== KÍCH THƯỚC DỮ LIỆU =====")
print("Số dòng, số cột:", df.shape)

# ===============================
# 2️⃣ KIỂM TRA DỮ LIỆU THIẾU
# ===============================
print("\n===== GIÁ TRỊ THIẾU =====")
print(df.isna().sum())

print("\n===== TỶ LỆ THIẾU (%) =====")
print((df.isna().mean() * 100).round(2))

# ===============================
# 3️⃣ THỐNG KÊ CƠ BẢN
# ===============================
print("\n===== THỐNG KÊ MÔ TẢ =====")
print(df.describe())

# ===============================
# 4️⃣ TÍNH TOÁN THỐNG KÊ RIÊNG LẺ
# ===============================
print("\n===== TRUNG BÌNH =====")
print(df.mean(numeric_only=True))

print("\n===== TỔNG =====")
print(df.sum(numeric_only=True))

print("\n===== MIN =====")
print(df.min(numeric_only=True))

print("\n===== MAX =====")
print(df.max(numeric_only=True))

print("\n===== ĐỘ LỆCH CHUẨN =====")
print(df.std(numeric_only=True))

# ===============================
# 5️⃣ ĐẾM DỮ LIỆU
# ===============================
print("\n===== SỐ DÒNG =====")
print(len(df))

print("\n===== ĐẾM GIÁ TRỊ KHÔNG THIẾU =====")
print(df.count())

print("\n===== SỐ GIÁ TRỊ DUY NHẤT =====")
print(df.nunique())

# ===============================
# 6️⃣ PHÂN TÍCH THEO NHÓM
# ===============================
print("\n===== TRUNG BÌNH THEO NĂM =====")
print(df.groupby("year").mean(numeric_only=True))

print("\n===== TỔNG THEO TỈNH =====")
print(df.groupby("province").sum(numeric_only=True))

print("\n===== ĐẾM SỐ BẢN GHI THEO TỈNH =====")
print(df.groupby("province").size())

# ===============================
# 7️⃣ NHIỀU THỐNG KÊ CÙNG LÚC
# ===============================
print("\n===== TỔNG HỢP THỐNG KÊ =====")
summary = df.groupby("year").agg({
    "population": ["mean", "sum", "min", "max"],
    "income": ["mean", "min", "max"],
    "unemployment_rate": ["mean", "max"]
})
print(summary)

# ===============================
# 8️⃣ THÔNG TIN QUAN TRỌNG (TÓM TẮT)
# ===============================
print("\n===== TÓM TẮT NHANH =====")
print("👉 Thu nhập trung bình:", df["income"].mean())
print("👉 Tổng dân số:", df["population"].sum())
print("👉 Tỷ lệ thất nghiệp cao nhất:", df["unemployment_rate"].max())
print("👉 Số tỉnh:", df["province"].nunique())
print("👉 Số năm:", df["year"].nunique())
```
