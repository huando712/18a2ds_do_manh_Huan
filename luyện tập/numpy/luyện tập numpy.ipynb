luyện theo file csv của thầy

```python
import numpy as np

# =============================
# Câu 1. Đọc dữ liệu từ file CSV
# =============================
data = np.genfromtxt(
    "data.csv",
    delimiter=",",
    dtype=None,
    encoding="utf-8",
    missing_values="",
    filling_values=np.nan
)

# =============================
# Câu 2. Tách các cột dữ liệu
# =============================
# data is a structured 1D array (fields f0..f6); access by field name
student_id = data['f0'].astype(str)
province = data['f1']
year = data['f2'].astype(int)
population = data['f3'].astype(float)
income = data['f4'].astype(float)
cpi = data['f5'].astype(float)
unemployment_rate = data['f6'].astype(float)
print(data)
# =============================
# Câu 3. Loại bỏ giá trị thiếu
# =============================
income_clean = income[~np.isnan(income)]
cpi_clean = cpi[~np.isnan(cpi)]
unemployment_clean = unemployment_rate[~np.isnan(unemployment_rate)]

# =============================
# Câu 4. Một số thống kê cơ bản
# =============================
print("Dân số lớn nhất:", np.max(population))
print("Thu nhập trung bình:", np.mean(income_clean))
print("CPI trung bình:", np.mean(cpi_clean))
print("Tỷ lệ thất nghiệp thấp nhất:", np.min(unemployment_clean))
```

GIẢI THÍCH CODE NUMPY (DẠNG TÓM TẮT – BÀI KIỂM TRA)

1. Import thư viện
import numpy as np
- Import NumPy, dùng bí danh np để gọi hàm cho ngắn gọn.

2. Đọc dữ liệu từ file CSV
data = np.genfromtxt(
    "data.csv",
    delimiter=",",
    dtype=None,
    encoding="utf-8",
    missing_values="",
    filling_values=np.nan
)

- genfromtxt: đọc dữ liệu từ file CSV
- delimiter=",": các cột ngăn cách bằng dấu phẩy
- dtype=None: NumPy tự nhận dạng kiểu dữ liệu
  → tạo structured array 1 chiều với các cột f0, f1, ..., f6
- encoding="utf-8": đọc được tiếng Việt
- missing_values="": xác định ô trống
- filling_values=np.nan: ô trống được gán NaN

3. Tách các cột dữ liệu
student_id = data['f0'].astype(str)
province = data['f1']
year = data['f2'].astype(int)
population = data['f3'].astype(float)
income = data['f4'].astype(float)
cpi = data['f5'].astype(float)
unemployment_rate = data['f6'].astype(float)

- Vì data là structured array 1D nên truy cập cột bằng data['f0']
- astype(): ép kiểu dữ liệu để có thể tính toán

4. Loại bỏ giá trị thiếu (NaN)
income_clean = income[~np.isnan(income)]
cpi_clean = cpi[~np.isnan(cpi)]
unemployment_clean = unemployment_rate[~np.isnan(unemployment_rate)]

- np.isnan(): kiểm tra giá trị NaN
- ~: phủ định, giữ lại các giá trị hợp lệ

5. Thống kê cơ bản
np.max(population): dân số lớn nhất
np.mean(income_clean): thu nhập trung bình
np.mean(cpi_clean): CPI trung bình
np.min(unemployment_clean): tỷ lệ thất nghiệp thấp nhất

KẾT LUẬN:
- dtype=None → structured array 1D
- Truy cập cột bằng data['f0'], không dùng data[:, i]
- Xử lý ô trống bằng NaN
- Dùng NumPy để lọc và thống kê dữ liệu
