- Viết class đơn giản để tổ chức các bước xử lý dữ liệu trong chương trình.
- Viết __init__ khởi tạo đối tượng của lớp để đọc file CSV
- Viết phương thứ đọc file CSV vào DataFrame
- Viết phương thức lọc dữ liệu theo student_id

```python
import numpy as np

class Economic_ontap:#Khai báo class Economic
    def __init__(self, file_path):#Hàm khởi tạo __init__
        self.file_path = file_path#đọc dữ liệu

    def load_data(self):#Phương thức load_data
        import pandas as pd
        return pd.read_csv(self.file_path)
    def filter_by_student(self, df, student_id):
        # Lọc dữ liệu theo student_id
        return df[df["student_id"] == student_id]#Trả về DataFrame chỉ chứa sinh viên đó
   
```
