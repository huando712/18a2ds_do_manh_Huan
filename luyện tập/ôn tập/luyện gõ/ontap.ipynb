2. NumPy – Xử lý giá trị thiếu (NaN)
Yêu cầu ôn tập:
- Trích xuất dữ liệu từ DataFrame sang mảng NumPy để thực hiện các phép toán số học.
- Phát hiện NaN trong mảng NumPy.
- Xử lý NaN bằng mean của cột tương ứng để đảm bảo dữ liệu hợp lệ cho phân tích.

```python
import numpy as np
arr=df["income"].to_numpy()#trích xuất dữ liệu sang numoy
mean_income=np.mean(arr)
arr[np.isnan(arr)]=mean_income

```
