PANDAS PHÂN TÍCH DỮ LIỆU

- MultiIndex (Thiết lập MultiIndex theo tỉnh/thành phố và năm để biểu diễn dữ liệu nhiều chiều.)
- groupby tổng hợp dữ liệu theo từng địa phương hoặc từng năm.
- pivot_table, tạo bảng tổng hợp phục vụ so sánh và phân tích xu hướng

```python
df_indexed = df_student.set_index(["province", "year"])
result = df_student.groupby("province")["income"].mean()
pivot_income = df_student.pivot_table(
    values="income",
    index="province",
    columns="year",
    aggfunc="mean"
)
```

SQLite – Hoàn thiện quy trình xử lý dữ liệu

- Lưu kết quả phân tích cuối cùng từ DataFrame vào cơ sở dữ liệu SQLite bằng hàm to_sql().
- Không lưu dữ liệu thô
Các hàm cần ôn:
- sqlite3.connect()
- DataFrame.to_sql()
Trong đó: hàm to_sql() thực hiện ghi (lưu) dữ liệu từ một DataFrame của Pandas vào một bảng nằm bên 
trong file CSDL SQLite (.db).

Kết nối cơ sở dữ liệu SQLite

```python
import sqlite3

conn = sqlite3.connect("ket_qua_phan_tich.db")
```

Lưu DataFrame vào SQLite bằng to_sql()

```python
pivot_income.to_sql(
    name="summary_income",
    con=conn,
    if_exists="replace",
    index=False
)
```
