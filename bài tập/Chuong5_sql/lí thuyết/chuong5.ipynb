```python
import sqlite3
# Kết nối đến cơ sở dữ liệu (hoặc tạo cơ sở dữ liệu mới nếu nó không tồn tại)
conn = sqlite3.connect('mydatabase.db')
# Tạo một đối tượng con trỏ (Cursor object) từ đối tượng kết nối
cursor = conn.cursor()
# Thực thi câu lệnh SQL và truy xuất dữ liệu được đặt trong cặp dấu '''...'''
cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
id INTEGER PRIMARY KEY, 
"Tên" TEXT, 
"Tuổi" INTEGER
)
''')
cursor.execute('''
INSERT INTO users ("Tên", "Tuổi") 
VALUES ("Nguyễn Văn A", 30)
''')
# Lưu các thay đổi vào cơ sở dữ liệu (nếu cần)
conn.commit()
# Truy xuất dữ liệu từ bảng users
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
# In kết quả
for row in rows:
print(row)
# Đóng con trỏ (Cursor object) và đóng kết nối
cursor.close()
conn.close()
```

```python
import sqlite3

with sqlite3.connect('example_db.db') as conn:
    cursor = conn.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS bang_diem (
        Name TEXT,
        Number REAL
    );
    """)

    # Ví dụ dữ liệu mẫu
    cursor.execute("INSERT INTO bang_diem (Name, Number) VALUES ('Anna', 8.5);")
    cursor.execute("INSERT INTO bang_diem (Name, Number) VALUES ('Binh', 9.0);")
    cursor.execute("INSERT INTO bang_diem (Name, Number) VALUES ('Nam', 6.8);")

    conn.commit()

print("✔ Bảng bang_diem đã được tạo và có dữ liệu mẫu.")
```

```python
import sqlite3

# 1. Kết nối CSDL
conn = sqlite3.connect('mydatabase.db')
cursor = conn.cursor()

# 2. Tạo bảng nếu chưa tồn tại
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    Ten TEXT,
    Tuoi INTEGER
);
""")

# 3. Nhập dữ liệu từ người dùng
name = input("Nhập tên người dùng: ")
age = int(input("Nhập tuổi người dùng: "))

# 4. Chèn dữ liệu vào bảng
cursor.execute("INSERT INTO users (Ten, Tuoi) VALUES (?, ?)", (name, age))
conn.commit()

# 5. In lại toàn bộ dữ liệu
print("\nDữ liệu trong bảng users:")
cursor = conn.execute("SELECT * FROM users")
for row in cursor:
    print(row)

# 6. Đóng kết nối
conn.close()
```

```python
import sqlite3
import matplotlib.pyplot as plt

# 1. Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('mydatabase.db')
cursor = conn.cursor()

# 2. Truy vấn dữ liệu (lấy cột Ten và Tuoi)
cursor.execute("SELECT Ten, Tuoi FROM users")
rows = cursor.fetchall()

# 3. Tách dữ liệu
Tên = [row[0] for row in rows]
Tuổi = [row[1] for row in rows]

# 4. In toàn bộ dữ liệu bảng users
print("Dữ liệu trong bảng users:")
cursor = conn.execute("SELECT * FROM users")
for row in cursor:
    print(row)

# 5. Đóng kết nối
conn.close()

# 6. Vẽ biểu đồ
plt.bar(Tên, Tuổi)
plt.xlabel('Tên')
plt.ylabel('Tuổi')
plt.title('Biểu đồ tuổi của người dùng')
plt.show()
```
