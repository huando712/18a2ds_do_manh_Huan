````python
import pandas as pd
#1. Định nghĩa Tuple đại diện cho tỉnh/TP - Năm tương ứng
index_tuples = [
                ('Hà Nội', 2015), ('Hà Nội', 2020),
                ('Nam Định', 2015), ('Nam Định', 2020),
                ('Thanh Hóa', 2015), ('Thanh Hóa', 2020),
                ('Đà Nẵng', 2015), ('Đà Nẵng', 2020),
                ('Hồ Chí Minh', 2015), ('Hồ Chí Minh', 2020)]
# 2. Dữ liệu (Sử dụng nguồn dữ liệu chính xác đã thống nhất)
data = [ 7216000, 8246540,
        1850600, 1780330,
        3514200, 3664940,
        1028800, 1169480,
        8146300, 9227600]
````

### ⚠️ `Phân tích sự bất cập của Phương pháp thủ công (Vòng lặp)`

#### Phương pháp trên kém hiệu quả và có các bất cập sau:

1.  **Bất cập về Hiệu suất (Vòng lặp)**: Để lọc dữ liệu, Pandas buộc phải **duyệt qua từng phần tử** trong Index. Đây là một **vòng lặp** (`for i in vn_pop.index`) và sẽ **cực kỳ chậm** với dữ liệu lớn (Big Data).
2.  **Bất cập về Tính rõ ràng:** Chúng ta phải biết chính xác **chỉ số 1** (`i[1]`) của Tuple để lấy thông tin "Năm". Nếu cấu trúc Index thay đổi (ví dụ: thêm cấp độ Tên quận), mã này sẽ **bị lỗi** vì chỉ số 1 không còn là "Năm" nữa.

#### Cách để tiếp cận và giải quyết các bất cập trên một cách đơn giản đó chính là **Hierarchical Indexing** thông qua **MultiIndex object**.


#### **Kết luận:** MultiIndex là giải pháp thay thế, cho phép chúng ta truy vấn theo **tên** (`'Năm'`) và sử dụng các thuật toán **được tối ưu hóa** của Pandas mà **không cần vòng lặp**.

#### **Ví dụ 4.6.1.4. Tạo MultiIndex từ danh sách tuples**

> #### Chỉ mục đa cấp theo chiều mở rộng – Multiplex as extra dimension[]
#### Phương thức **unstack()**: Chuyển đổi một Series có chỉ mục đa cấp thành một DataFrame thông thường. 
#### Cú pháp:
``` Python
        Series.unstack(level=-1, fill_value=None)
```
Trong đó:
  - level (tùy chọn): Xác định cấp chỉ mục muốn di chuyển từ hàng thành cột. Mặc định là -1, tức là di chuyển cấp chỉ mục cuối cùng.
  - fill_value (tùy chọn): Xác định giá trị được sử dụng để điền vào các ô bị thiếu sau khi unstack. Mặc định là None, nghĩa là các ô thiếu sẽ được điền bằng giá trị NaN.
#### Lưu ý: áp dụng cho đối tượng Series, trong đó Series phải có chỉ mục đa cấp.



#### **Ví dụ 4.6.1.7–4.6.1.8. Chuyển đổi giữa dạng rộng và dài (unstack/stack)**
Tạo **DataFrame** từ `vn_pop`, sau đó dùng **unstack** để đưa **Năm** lên cột (rộng), và **stack** để quay lại dạng **Series MultiIndex** (dài).


````python
vn_pop_mul              #Series đa cấp
````

#### Cú pháp:
```Python
    DataFrame.stack(level=-1, dropna=True)
```
Trong đó:
  - level là một tham số tùy chọn để chỉ định mức chỉ mục (level) hoặc tên cột để thực hiện stack. Mặc định là -1, có nghĩa là stack tất cả các mức chỉ mục hiện có.
  - dropna là một tham số tùy chọn để chỉ định xem liệu các giá trị NaN (không có giá trị) có nên được bỏ qua hay không. Mặc định là True, nghĩa là các hàng chứa giá trị NaN sẽ không được bao gồm trong kết quả.

#### **e. Tạo MultiIndex cho cột**
#### **Ví dụ 4.6.2.5**. Tạo một DataFrame với một chỉ mục đa cấp cho cả hàng và cột để lưu trữ dữ liệu điểm thi các môn Toán, Lý, Hóa của các học sinh, theo năm học, đợt thi, môn thi như sau:

````python
import pandas as pd
import numpy as np
index=pd.MultiIndex.from_product([[2020,2021,2022],[1,2,3]],                    # 3 năm học 2020, 2021, 2022
                          names=['Năm','Lần thi'])                              # Lần thi : 1, 2, 3
columns=pd.MultiIndex.from_product([['Bình','Lan','Vân'],
                                 ['Toán','Lý','Hóa']],
                           names=['Học sinh','Môn học'])

#Tạo dữ liệu mẫu, là mảng kích thước 9x9 với các giá trị ngẫu nhiên từ 5-10
data=np.random.randint(5,10,(9,9))
'''Tạo DataFrame score với dữ liệu ngẫu nhiên từ mảng data kích thước 9x9 trong đó các chỉ mục hàng và cột được thiết lập'''
score=pd.DataFrame(data,index=index, columns=columns)
score
````

#### **4.6.3.2.Cắt nhanh một cấp độ (.xs(level=...))**
#### Sử dụng .xs() để chọn một giá trị duy nhất tại một cấp độ Index cụ thể, loại bỏ cấp độ đó khỏi kết quả (mặc định).

#### **Phương thức .xs()** cho phép thực hiện một "lát cắt ngang" (cross-section) của dữ liệu. Nó được dùng để chọn một giá trị duy nhất từ một cấp độ Index cụ thể, đồng thời loại bỏ cấp độ đó ra khỏi Index kết quả (mặc định).
#### Cú pháp:
```python
            DataFrame.xs(key, axis=0, level=None, drop_level=True)
    trong đó:
```
* #### **key**: Giá trị cụ thể  muốn chọn (ví dụ: 2021, 'Toán').
* #### **axis**: Trục muốn áp dụng: 0 (Index/Hàng) hoặc 1 (Columns/Cột).
* #### **level**: Tên hoặc vị trí (số nguyên) của cấp độ Index muốn cắt.
* #### **drop_level**: Mặc định là True. Nếu True, cấp độ Index bạn vừa dùng để cắt sẽ bị loại bỏ khỏi Index của kết quả trả về.



#### **Ví dụ** Sử dụng dataFrame score ở trên
* #### Index hàng: (Năm, Lần thi) 
* #### Index cột: (Học sinh, Môn học)

````python
score
````

````python
# Lấy kết quả Lần 3 của tất cả các năm
print(score.xs(3, level='Lần thi'))
````

* #### .mean(): tính trung bình trên từng nhóm cột.
