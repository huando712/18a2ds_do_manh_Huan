Bài 4.1. Tạo Series từ list, numpy và dict

```python
import numpy as np
import pandas as pd

mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

# Tạo các Series
ser1 = pd.Series(mylist)
ser2 = pd.Series(myarr)
ser3 = pd.Series(mydict)
```

Bài 4.3. Kết nối các Series thành DataFrame

```python
ser1 = pd.Series(list('abcedfghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))

# Cách 1: Kết nối theo cột (axis=1)
df = pd.concat([ser1, ser2], axis=1)

# Cách 2: Dùng dictionary
df = pd.DataFrame({'col1': ser1, 'col2': ser2})
```

Bài 4.4. Thao tác so sánh giữa 2 Series

```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

# 1. Xóa các mục của ser1 có mặt trong ser2
res1 = ser1[~ser1.isin(ser2)]

# 2. Lấy các phần tử không nằm chung trong cả hai (Phần bù đối xứng)
union = pd.Series(np.union1d(ser1, ser2))
intersect = pd.Series(np.intersect1d(ser1, ser2))
res2 = union[~union.isin(intersect)]
```

Bài 4.5. Tính toán thống kê cơ bản

```python
ser = pd.Series(np.random.normal(10, 5, 25))

# Sử dụng hàm describe() hoặc các hàm riêng lẻ
stats = {
    'min': ser.min(),
    '25th': ser.quantile(0.25),
    'median': ser.median(),
    '75th': ser.quantile(0.75),
    'max': ser.max()
}
```

Bài 4.7. Giữ lại 2 giá trị xuất hiện nhiều nhất

```python
np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

# Tìm top 2 giá trị xuất hiện nhiều nhất
top2 = ser.value_counts().index[:2]

# Thay thế các giá trị còn lại bằng 'Other'
ser[~ser.isin(top2)] = 'Other'
```

Bài 4.8. Chia chuỗi thành các phân vị (Deciles)

```python
ser = pd.Series(np.random.random(20))

# Chia làm 10 phần bằng nhau và gán nhãn
ser_bins = pd.qcut(ser, q=10, labels=['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th', '10th'])
```
