```python
import pandas as pd
import numpy as np

# 1. Tạo bộ sinh số ngẫu nhiên cố định (seed = 100)
rng = np.random.RandomState(100)

# 2. Tạo mảng NumPy ngẫu nhiên
arr1 = rng.randint(0, 10, 4)       # Mảng 1 chiều gồm 4 số ngẫu nhiên [0,9]
arr2 = rng.randint(0, 10, (3, 4))  # Mảng 2 chiều (3x4) gồm số ngẫu nhiên [0,9]

# 3. Tạo Series và DataFrame từ các mảng
sr = pd.Series(arr1)                            # Series 1 chiều
df = pd.DataFrame(arr2, columns=['A','B','C','D'])  # DataFrame có 4 cột

# 4. In kết quả
print("Series:\n", sr)
print("DataFrame:\n", df)
```

```python
import pandas as pd

# Số liệu top 5 tỉnh/thành đông dân nhất Việt Nam (2023)
population = pd.Series({
    'TP.Hồ Chí Minh': 9411805,
    'Hà Nội': 8330830,
    'Thanh Hóa': 3720000,
    'Nghệ An': 3410000,
    'Đồng Nai': 3097107
})

# Số liệu top 5 tỉnh/thành có diện tích lớn nhất Việt Nam (2023)
area = pd.Series({
    'Nghệ An': 16493,
    'Gia Lai': 15536.9,
    'Sơn La': 14174.4,
    'Đăk Lăk': 13030.5,
    'Thanh Hóa': 11132.2
})

# In kết quả
print("Top 5 tỉnh/thành đông dân nhất Việt Nam 2023:\n", population)
print("\nTop 5 tỉnh/thành có diện tích lớn nhất Việt Nam 2023:\n", area)
density = area/population
print("Mật độ dân số tỉnh/thành :\n", density)
```

```python
import pandas as pd

data = pd.DataFrame({
    'Dân số': [8330830, 1320000, 1880000, 1220000],
    'Diện tích': [3358.6, 6177.7, 1667.7, 617.0],
    'Mật độ': [2480, 213, 1127, 1977]
}, index=['Hà Nội', 'Quảng Ninh', 'Nam Định', 'Ninh Bình'])

print("Bảng dữ liệu gốc:\n", data)
```

```python
print(data)
#Lấy hàng đầu tiên
print("\nDữ liệu của thủ đô Hà Nội:\n",data.iloc[0])
#Lấy dữ liệu 3 hàng đầu tiên
print("\nDữ liệu của Hà Nội, Quảng Ninh và Nam Định:\n",data.iloc[0:3])
#Lấy dữ liệu 2 hàng, 2 cột đầu tiên
print("\nDữ liệu dân số và diện tích của Hà Nội và Quảng Ninh:\n",data.iloc[:2,:2])
#Lấy dữ liệu về mật độ dân số của các tỉnh còn lại
print("Mật độ dân số của các tỉnh/thành phố:\n",data.iloc[:,-1])
```
