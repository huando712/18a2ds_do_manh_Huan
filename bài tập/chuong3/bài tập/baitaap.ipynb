Bài 3.2: Thao tác với mảng cơ bản

```python
# 1. Tạo mảng từ 0-9
arr = np.arange(10)
print("Mảng arr:", arr)
print("Kiểu dữ liệu:", arr.dtype, "| Kích thước:", arr.shape)

# 2. Tạo arr_odd và arr_even
arr_odd = arr[arr % 2 != 0]
arr_even = arr[arr % 2 == 0]
print("Lẻ:", arr_odd, "| Chẵn:", arr_even)

# 3. Thay phần tử lẻ bằng 100
arr_update_1 = arr.copy()
arr_update_1[arr_update_1 % 2 != 0] = 100
print("Mảng sau khi cập nhật:", arr_update_1)
```

Bài 3.3: So sánh và lọc mảng

```python
arr_a = np.array([1,2,3,2,3,4,3,4,5,6])
arr_b = np.array([7,2,10,2,7,4,9,4,9,8])

# 1. Phần tử xuất hiện ở cả a và b (giao nhau)
arr_c = np.intersect1d(arr_a, arr_b)
print("Phần tử chung:", arr_c)

# 2. Phần tử chỉ xuất hiện ở a (hiệu của a và b)
arr_d = np.setdiff1d(arr_a, arr_b)
print("Chỉ có ở a:", arr_d)

# 3. Lọc phần tử từ 5 đến 10
arr_e = np.array([2, 6, 1, 9, 10, 3, 27, 8, 6, 25, 16])
arr_f = arr_e[(arr_e >= 5) & (arr_e <= 10)]
print("Phần tử từ 5 đến 10:", arr_f)
```

Bài 3.4: Thêm, xóa và cập nhật mảng

```python
# 1. Tạo arr_zeros và cập nhật vị trí thứ 5 (index 4) là 1
# Lưu ý: "Vị trí thứ 5" thường là index 4, nhưng nếu đề hiểu theo index 5 thì sửa lại là [5]
arr_zeros = np.zeros(10)
arr_zeros[4] = 1 
print("arr_zeros:", arr_zeros)

# 2. Mảng từ 10-24 và đảo ngược
arr_h = np.arange(10, 25)
print("Đảo ngược:", arr_h[::-1])

# 3. Lọc các phần tử khác 0
arr_k = np.array([1, 2, 0, 8, 2, 0, 1, 3, 0, 5, 0])
arr_l = arr_k[arr_k != 0]
print("Khác 0:", arr_l)

# 4. Thêm 10 và 20 vào cuối mảng
arr_l = np.append(arr_l, [10, 20])

# 5. Thêm 100 vào vị trí index = 5
arr_l = np.insert(arr_l, 5, 100)

# 6. Xóa các phần tử tại vị trí index 0, 1, 2
arr_l = np.delete(arr_l, [0, 1, 2])
print("Mảng cuối cùng bài 3.4:", arr_l)
```
