BÀI 2.1 – Ghi dữ liệu quản lý giao dịch vào file JSON

```python
import json
from datetime import datetime

def nhap_giao_dich():
    ds = []
    while True:
        loai = input("Loại giao dịch (tien/vang): ")
        so_tien = float(input("Số tiền: "))
        ds.append({
            "loai": loai,
            "so_tien": so_tien
        })
        cont = input("Tiếp tục giao dịch? (y/n): ")
        if cont.lower() != 'y':
            break
    return ds

giao_dich = nhap_giao_dich()
chon = input("Ghi vào file? (1: Có, 0: Không): ")

if chon == "1":
    now = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    filename = f"{now}.json"
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(giao_dich, f, ensure_ascii=False, indent=4)
    print("Đã ghi file:", filename)
else:
    print("Không ghi file")
```

```python
from xml.dom import minidom

doc = minidom.parse("sample.xml")
staffs = doc.getElementsByTagName("staff")

for s in staffs:
    name = s.getElementsByTagName("name")[0].firstChild.data
    salary = s.getElementsByTagName("salary")[0].firstChild.data
    print(name, "-", salary)
```

BÀI 2.6 – Đọc RSS Feed, phân tích XML, lưu CSV

```python
import requests
import xml.etree.ElementTree as ET
import csv

url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"
xml_data = requests.get(url).content

with open("rss.xml", "wb") as f:
    f.write(xml_data)

root = ET.fromstring(xml_data)

items = []
for item in root.findall(".//item"):
    title = item.find("title").text
    link = item.find("link").text
    items.append({"title": title, "link": link})

with open("news.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "link"])
    writer.writeheader()
    writer.writerows(items)

print("Đã lưu CSV")
```

BÀI 2.9 – Dictionary → JSON (sắp xếp key)

```python
{
  "company": "Công ty TNHH Đại Việt",
  "address": "ABC Giải Phóng - Hà Nội",
  "employees": [
    {"name": "A", "unit": "A1"},
    {"name": "B", "unit": "A2"},
    {"name": "C", "unit": "A1"},
    {"name": "D", "unit": "A3"}
  ]
}
```

```python
import json
from collections import Counter

with open("employees.json", "r", encoding="utf-8") as f:
    data = json.load(f)

emps = data["employees"]
total = len(emps)

counter = Counter(e["unit"] for e in emps)

print("Tên công ty:", data["company"])
print("Địa chỉ:", data["address"])
print("Tổng số nhân viên:", total)

for unit, count in counter.items():
    print(f"Đơn vị {unit}: {count} nhân viên - {count/total*100:.2f}%")
```
