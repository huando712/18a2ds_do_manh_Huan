```python
# VÃ­ dá»¥: Deserialize - chuyá»ƒn chuá»—i JSON thÃ nh Ä‘á»‘i tÆ°á»£ng Python (cÃ³ xá»­ lÃ½ sá»‘ phá»©c)
import json

# HÃ m kiá»ƒm tra chuá»—i JSON cÃ³ chá»©a Ä‘á»‘i tÆ°á»£ng phá»©c (complex object) hay khÃ´ng
def is_complex(object):
    # Kiá»ƒm tra xem trong Ä‘á»‘i tÆ°á»£ng cÃ³ tá»“n táº¡i khÃ³a '__complex__' hay khÃ´ng
    if '__complex__' in object:
        # Náº¿u cÃ³, táº¡o má»™t sá»‘ phá»©c tá»« thÃ´ng tin trong Ä‘á»‘i tÆ°á»£ng JSON
        return complex(object['real'], object['img'])
    # Náº¿u khÃ´ng cÃ³ khÃ³a '__complex__', Ä‘Ã¢y khÃ´ng pháº£i Ä‘á»‘i tÆ°á»£ng phá»©c â†’ tráº£ vá» Ä‘á»‘i tÆ°á»£ng gá»‘c
    return object


def main():
    # 1ï¸âƒ£ Sá»­ dá»¥ng json.loads() vá»›i object_hook Ä‘á»ƒ kiá»ƒm tra vÃ  chuyá»ƒn Ä‘á»•i Ä‘á»‘i tÆ°á»£ng
    # VÃ­ dá»¥ 1: chuá»—i JSON mÃ´ táº£ má»™t sá»‘ phá»©c (cÃ³ khÃ³a "__complex__")
    complex_object = json.loads(
        '{"__complex__": true, "real": 4, "img": 5}',
        object_hook=is_complex
    )

    # VÃ­ dá»¥ 2: chuá»—i JSON bÃ¬nh thÆ°á»ng (khÃ´ng cÃ³ khÃ³a "__complex__")
    simple_object = json.loads(
        '{"real": 6, "img": 7}',
        object_hook=is_complex
    )

    # 2ï¸âƒ£ In káº¿t quáº£ Ä‘á»ƒ so sÃ¡nh
    print("Äá»‘i tÆ°á»£ng sá»‘ phá»©c:", complex_object)
    print("Äá»‘i tÆ°á»£ng bÃ¬nh thÆ°á»ng:", simple_object)


# 3ï¸âƒ£ Cháº¡y hÃ m main() khi chÆ°Æ¡ng trÃ¬nh Ä‘Æ°á»£c thá»±c thi trá»±c tiáº¿p
if __name__ == "__main__":
    main()
```

```python
# VÃ­ dá»¥ 2.2.2.2.1: Minh há»a Deserialize sá»­ dá»¥ng object_hook
# DÃ¹ng Ä‘á»ƒ chuyá»ƒn chuá»—i JSON chá»©a Ä‘á»‘i tÆ°á»£ng phá»©c (complex object) thÃ nh Ä‘á»‘i tÆ°á»£ng Python

import json

# HÃ m kiá»ƒm tra vÃ  xá»­ lÃ½ náº¿u JSON chá»©a Ä‘á»‘i tÆ°á»£ng phá»©c
def is_complex(object):
    # Kiá»ƒm tra xem trong Ä‘á»‘i tÆ°á»£ng cÃ³ tá»“n táº¡i khÃ³a '__complex__' hay khÃ´ng
    if '__complex__' in object:
        # Náº¿u cÃ³, táº¡o ra má»™t sá»‘ phá»©c tá»« cÃ¡c thÃ nh pháº§n 'real' vÃ  'img'
        return complex(object['real'], object['img'])
    # Náº¿u khÃ´ng cÃ³, tráº£ vá» Ä‘á»‘i tÆ°á»£ng gá»‘c (dict bÃ¬nh thÆ°á»ng)
    return object

def main():
    # 1ï¸âƒ£ Chuá»—i JSON mÃ´ táº£ má»™t sá»‘ phá»©c (cÃ³ khÃ³a "__complex__")
    json_complex = '{"__complex__": true, "real": 4, "img": 5}'
    
    # 2ï¸âƒ£ Chuá»—i JSON mÃ´ táº£ Ä‘á»‘i tÆ°á»£ng bÃ¬nh thÆ°á»ng (khÃ´ng cÃ³ khÃ³a "__complex__")
    json_simple = '{"real": 6, "img": 7}'
    
    # 3ï¸âƒ£ Sá»­ dá»¥ng json.loads() kÃ¨m object_hook Ä‘á»ƒ tá»± Ä‘á»™ng xá»­ lÃ½ Ä‘á»‘i tÆ°á»£ng
    complex_object = json.loads(json_complex, object_hook=is_complex)
    simple_object = json.loads(json_simple, object_hook=is_complex)
    
    # 4ï¸âƒ£ In káº¿t quáº£ ra mÃ n hÃ¬nh
    print("Äá»‘i tÆ°á»£ng sá»‘ phá»©c:", complex_object)
    print("Äá»‘i tÆ°á»£ng bÃ¬nh thÆ°á»ng:", simple_object)

# 5ï¸âƒ£ Khi cháº¡y trá»±c tiáº¿p file, gá»i hÃ m main()
if __name__ == "__main__":
    main()
```

```python
import json
from decimal import Decimal  # cung cáº¥p lá»›p Decimal cho phÃ©p tÃ­nh chÃ­nh xÃ¡c cao vá»›i sá»‘ tháº­p phÃ¢n

# 1ï¸âƒ£ Chuá»—i JSON gá»‘c
json_string = '{"name": "Alice", "age": 25, "balance": 1000.5, "status": NaN}'

# 2ï¸âƒ£ PhÃ¢n tÃ­ch chuá»—i JSON báº±ng json.loads() vá»›i cÃ¡c hÃ m chuyá»ƒn Ä‘á»•i tÃ¹y chá»‰nh
data = json.loads(
    json_string,
    parse_float=Decimal,            # Chuyá»ƒn cÃ¡c giÃ¡ trá»‹ float (1000.5) thÃ nh kiá»ƒu Decimal
    parse_int=str,                  # Chuyá»ƒn cÃ¡c giÃ¡ trá»‹ int (25) thÃ nh chuá»—i "25"
    parse_constant=lambda x: None   # Chuyá»ƒn cÃ¡c giÃ¡ trá»‹ Ä‘áº·c biá»‡t nhÆ° NaN, Infinity thÃ nh None
)

# 3ï¸âƒ£ In káº¿t quáº£ sau khi chuyá»ƒn Ä‘á»•i
print(data)
```

```python
import json

# ğŸ§© 1. Äá»‹nh nghÄ©a má»™t lá»›p Ä‘á»‘i tÆ°á»£ng tÃ¹y chá»‰nh
class CustomObject:
    def __init__(self, data):
        self.data = data

    def to_dict(self):
        # HÃ m chuyá»ƒn Ä‘á»•i Ä‘á»‘i tÆ°á»£ng CustomObject thÃ nh dict (Ä‘á»ƒ JSON hiá»ƒu Ä‘Æ°á»£c)
        return {"custom_data": self.data}


# ğŸ§  2. Táº¡o má»™t lá»›p JSONEncoder tÃ¹y chá»‰nh Ä‘á»ƒ xá»­ lÃ½ Ä‘á»‘i tÆ°á»£ng CustomObject
class CustomObjectEncoder(json.JSONEncoder):
    def default(self, obj):
        # Kiá»ƒm tra náº¿u obj lÃ  kiá»ƒu CustomObject thÃ¬ chuyá»ƒn sang dict
        if isinstance(obj, CustomObject):
            return obj.to_dict()
        # Náº¿u khÃ´ng pháº£i CustomObject thÃ¬ gá»i phÆ°Æ¡ng thá»©c máº·c Ä‘á»‹nh cá»§a JSONEncoder
        return super().default(obj)


# ğŸ’¡ 3. Táº¡o má»™t Ä‘á»‘i tÆ°á»£ng CustomObject
custom_obj = CustomObject({'key': 'value'})

# ğŸ’¾ 4. Sá»­ dá»¥ng JSONEncoder tÃ¹y chá»‰nh Ä‘á»ƒ mÃ£ hÃ³a Ä‘á»‘i tÆ°á»£ng CustomObject
json_data = json.dumps(custom_obj, cls=CustomObjectEncoder, ensure_ascii=False, indent=4)

# ğŸ–¨ï¸ 5. In káº¿t quáº£ ra mÃ n hÃ¬nh
print(json_data)
```

```python
import json
# Import lá»›p JSONDecoder tá»« module json.decoder
from json.decoder import JSONDecoder

# Chuá»—i JSON chá»©a dá»¯ liá»‡u mÃ u sáº¯c
colour_string = '{"colour": ["red", "yellow"]}'

# ğŸ§  Sá»­ dá»¥ng JSONDecoder Ä‘á»ƒ giáº£i mÃ£ chuá»—i JSON thÃ nh Ä‘á»‘i tÆ°á»£ng Python.
# JSONDecoder lÃ  lá»›p con cá»§a json.JSONDecoder, cho phÃ©p tÃ¹y chá»‰nh quÃ¡ trÃ¬nh giáº£i mÃ£.
# PhÆ°Æ¡ng thá»©c decode() sáº½ giáº£i mÃ£ chuá»—i JSON vÃ  tráº£ vá» Ä‘á»‘i tÆ°á»£ng Python tÆ°Æ¡ng á»©ng.
decoded_object = JSONDecoder().decode(colour_string)

# ğŸ–¨ï¸ In Ä‘á»‘i tÆ°á»£ng Python Ä‘Ã£ giáº£i mÃ£ vÃ  kiá»ƒu dá»¯ liá»‡u cá»§a Ä‘á»‘i tÆ°á»£ng Ä‘Ã³
print(decoded_object)
print(type(decoded_object))
```
