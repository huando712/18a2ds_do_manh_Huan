ưu điểm nhược điểm json

Ưu điểm:
 Định dạng đơn giản: JSON được thiết kế để đọc và viết một cách dễ dàng cho con người, đây là một
định dạng dữ liệu đơn giản, dễ hiểu và đọc.
 Hỗ trợ nhiều ngôn ngữ lập trình: JSON được hỗ trợ bởi hầu hết các ngôn ngữ lập trình phổ biến như
JavaScript, Python, Ruby, PHP và nhiều ngôn ngữ khác.
 Kích thước nhỏ: JSON có kích thước nhỏ hơn so với một số định dạng dữ liệu khác, giúp cho việc
truyền tải và lưu trữ dữ liệu dễ dàng hơn.
 Dữ liệu có cấu trúc: JSON sử dụng cấu trúc dữ liệu key-value, mảng và đối tượng, cho phép xử lý dữ
liệu phức tạp hơn.
 Hỗ trợ tương tác với các API: JSON được sử dụng rộng rãi để truyền dữ liệu giữa các ứng dụng và
tương tác với các API (Application Programming Interface).
Nhược điểm
 Không có hỗ trợ cho các tính năng chuyên biệt: JSON không cung cấp các tính năng chuyên biệt như
kiểm tra lỗi, bảo mật dữ liệu, nén dữ liệu, v.v.
 Không hỗ trợ các kiểu dữ liệu phức tạp: JSON không hỗ trợ các kiểu dữ liệu phức tạp như ngày giờ, số
thập phân, v.v. nên trong một số trường hợp, nó có thể gây khó khăn trong việc lưu trữ và truyền tải
dữ liệu.
 Dữ liệu chỉ có thể được sử dụng khi được giải mã: Dữ liệu JSON cần được giải mã bởi ngôn ngữ lập
trình trước khi có thể sử dụng. Việc này có thể tốn nhiều thời gian và tài nguyên hệ thống nếu dữ liệu
lớn hoặc phức tạp.

```python
# Ví dụ 2.2.2.2b: Minh họa Serialize
# Chuyển đổi đối tượng Python (data) thành chuỗi JSON và ghi vào file thongtin.json

import json  # Nhập thư viện JSON có sẵn trong Python

# 1️⃣ Tạo đối tượng Python (dạng dict)
data = {
    "Ho_va_ten": "Nguyễn Văn A",
    "Nam_sinh": 2025,
    "Gioi_tinh": "Nam",
    "so_dt": "0123456789",
    "email": "nguyenvana@uneti.edu.vn"
}

# 2️⃣ Ghi dữ liệu vào file JSON
with open('thongtin.json', 'w', encoding='utf-8') as f:
    # Ghi đối tượng Python vào file dưới dạng JSON
    json.dump(data, f, ensure_ascii=False, indent=4)

print("✅ Đã ghi dữ liệu thành công vào tệp 'thongtin.json'.")

# 3️⃣ Đọc lại file để kiểm tra nội dung
with open('thongtin.json', 'r', encoding='utf-8') as f:
    print("\n--- Nội dung tệp 'thongtin.json' ---")
    print(f.read())
```

```python
# Ví dụ: Deserialize - chuyển chuỗi JSON thành đối tượng Python (có xử lý số phức)
import json

# Hàm kiểm tra chuỗi JSON có chứa đối tượng phức (complex object) hay không
def is_complex(object):
    # Kiểm tra xem trong đối tượng có tồn tại khóa '__complex__' hay không
    if '__complex__' in object:
        # Nếu có, tạo một số phức từ thông tin trong đối tượng JSON
        return complex(object['real'], object['img'])
    # Nếu không có khóa '__complex__', đây không phải đối tượng phức → trả về đối tượng gốc
    return object


def main():
    # 1️⃣ Sử dụng json.loads() với object_hook để kiểm tra và chuyển đổi đối tượng
    # Ví dụ 1: chuỗi JSON mô tả một số phức (có khóa "__complex__")
    complex_object = json.loads(
        '{"__complex__": true, "real": 4, "img": 5}',
        object_hook=is_complex
    )

    # Ví dụ 2: chuỗi JSON bình thường (không có khóa "__complex__")
    simple_object = json.loads(
        '{"real": 6, "img": 7}',
        object_hook=is_complex
    )

    # 2️⃣ In kết quả để so sánh
    print("Đối tượng số phức:", complex_object)
    print("Đối tượng bình thường:", simple_object)


# 3️⃣ Chạy hàm main() khi chương trình được thực thi trực tiếp
if __name__ == "__main__":
    main()
```

sử dụng object


```python
# Ví dụ 2.2.2.2.1: Minh họa Deserialize sử dụng object_hook
# Dùng để chuyển chuỗi JSON chứa đối tượng phức (complex object) thành đối tượng Python

import json

# Hàm kiểm tra và xử lý nếu JSON chứa đối tượng phức
def is_complex(object):
    # Kiểm tra xem trong đối tượng có tồn tại khóa '__complex__' hay không
    if '__complex__' in object:
        # Nếu có, tạo ra một số phức từ các thành phần 'real' và 'img'
        return complex(object['real'], object['img'])
    # Nếu không có, trả về đối tượng gốc (dict bình thường)
    return object

def main():
    # 1️⃣ Chuỗi JSON mô tả một số phức (có khóa "__complex__")
    json_complex = '{"__complex__": true, "real": 4, "img": 5}'
    
    # 2️⃣ Chuỗi JSON mô tả đối tượng bình thường (không có khóa "__complex__")
    json_simple = '{"real": 6, "img": 7}'
    
    # 3️⃣ Sử dụng json.loads() kèm object_hook để tự động xử lý đối tượng
    complex_object = json.loads(json_complex, object_hook=is_complex)
    simple_object = json.loads(json_simple, object_hook=is_complex)
    
    # 4️⃣ In kết quả ra màn hình
    print("Đối tượng số phức:", complex_object)
    print("Đối tượng bình thường:", simple_object)

# 5️⃣ Khi chạy trực tiếp file, gọi hàm main()
if __name__ == "__main__":
    main()
```

```python
import json
from decimal import Decimal  # cung cấp lớp Decimal cho phép tính chính xác cao với số thập phân

# 1️⃣ Chuỗi JSON gốc
json_string = '{"name": "Alice", "age": 25, "balance": 1000.5, "status": NaN}'

# 2️⃣ Phân tích chuỗi JSON bằng json.loads() với các hàm chuyển đổi tùy chỉnh
data = json.loads(
    json_string,
    parse_float=Decimal,            # Chuyển các giá trị float (1000.5) thành kiểu Decimal
    parse_int=str,                  # Chuyển các giá trị int (25) thành chuỗi "25"
    parse_constant=lambda x: None   # Chuyển các giá trị đặc biệt như NaN, Infinity thành None
)

# 3️⃣ In kết quả sau khi chuyển đổi
print(data)
```

```python
import json

# 🧩 1. Định nghĩa một lớp đối tượng tùy chỉnh
class CustomObject:
    def __init__(self, data):
        self.data = data

    def to_dict(self):
        # Hàm chuyển đổi đối tượng CustomObject thành dict (để JSON hiểu được)
        return {"custom_data": self.data}


# 🧠 2. Tạo một lớp JSONEncoder tùy chỉnh để xử lý đối tượng CustomObject
class CustomObjectEncoder(json.JSONEncoder):
    def default(self, obj):
        # Kiểm tra nếu obj là kiểu CustomObject thì chuyển sang dict
        if isinstance(obj, CustomObject):
            return obj.to_dict()
        # Nếu không phải CustomObject thì gọi phương thức mặc định của JSONEncoder
        return super().default(obj)


# 💡 3. Tạo một đối tượng CustomObject
custom_obj = CustomObject({'key': 'value'})

# 💾 4. Sử dụng JSONEncoder tùy chỉnh để mã hóa đối tượng CustomObject
json_data = json.dumps(custom_obj, cls=CustomObjectEncoder, ensure_ascii=False, indent=4)

# 🖨️ 5. In kết quả ra màn hình
print(json_data)
```

```python
import json
# Import lớp JSONDecoder từ module json.decoder
from json.decoder import JSONDecoder

# Chuỗi JSON chứa dữ liệu màu sắc
colour_string = '{"colour": ["red", "yellow"]}'

# 🧠 Sử dụng JSONDecoder để giải mã chuỗi JSON thành đối tượng Python.
# JSONDecoder là lớp con của json.JSONDecoder, cho phép tùy chỉnh quá trình giải mã.
# Phương thức decode() sẽ giải mã chuỗi JSON và trả về đối tượng Python tương ứng.
decoded_object = JSONDecoder().decode(colour_string)

# 🖨️ In đối tượng Python đã giải mã và kiểu dữ liệu của đối tượng đó
print(decoded_object)
print(type(decoded_object))
```

```python
import requests

# Thực hiện yêu cầu HTTP đến URL 'https://jsonplaceholder.typicode.com/posts'
response = requests.get('https://jsonplaceholder.typicode.com/posts')

# Kiểm tra xem yêu cầu thành công hay không (mã trạng thái 200 = OK)
if response.status_code == 200:
    # Nếu yêu cầu thành công, lấy dữ liệu JSON từ phản hồi
    json_data = response.json()

    # In ra màn hình thông tin các bài đăng từ dữ liệu JSON
    for post in json_data:
        print("ID bài đăng:", post['id'])
        print("Tiêu đề:", post['title'])
        print("Nội dung:", post['body'])
        print("==================================")

else:
    # Nếu không thành công, in ra mã lỗi HTTP
    print("Yêu cầu không thành công. Mã trạng thái:", response.status_code)
```
