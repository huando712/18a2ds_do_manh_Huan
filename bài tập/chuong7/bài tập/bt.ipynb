```python
import tkinter as tk
from tkinter import messagebox, Menu
from tkinter import ttk
import math

# --- CẤU HÌNH CHUNG ---
FONT_LABEL = ("Arial", 12)
FONT_ENTRY = ("Arial", 12)
PADDING = 10

class GiaiBaiTapChuong7:
    def __init__(self, root):
        self.root = root
        self.root.title("Giải Bài Tập Python - Chương 7 (GUI)")
        self.root.geometry("600x600")
        
        # Tiêu đề chính
        lbl_title = tk.Label(root, text="CHƯƠNG 7: LẬP TRÌNH GUI VỚI TKINTER", 
                             font=("Arial", 16, "bold"), fg="blue")
        lbl_title.pack(pady=20)

        # Khung chứa các nút
        frame_btn = tk.Frame(root)
        frame_btn.pack(fill="both", expand=True, padx=20, pady=10)

        # Danh sách bài tập
        exercises = [
            ("Bài 7.1: Cửa sổ đơn giản", self.bai_7_1),
            ("Bài 7.2: Nhãn (Label)", self.bai_7_2),
            ("Bài 7.3: Định dạng Font", self.bai_7_3),
            ("Bài 7.4: Form nhập liệu", self.bai_7_4),
            ("Bài 7.5: Hiển thị ảnh", self.bai_7_5),
            ("Bài 7.6: Đảo ngược chuỗi", self.bai_7_6),
            ("Bài 7.7: Tính tổng 1..N", self.bai_7_7),
            ("Bài 7.8: Tìm ước số", self.bai_7_8),
            ("Bài 7.9: GCD và LCM", self.bai_7_9),
            ("Bài 7.10: Kiểm tra tuổi", self.bai_7_10),
            ("Bài 7.11: Đổi Dương lịch -> Âm lịch", self.bai_7_11),
            ("Bài 7.12: Tính BMI", self.bai_7_12),
            ("Bài 7.13: Tạo Menu", self.bai_7_13),
            ("Bài 7.14: Tính cước Taxi", self.bai_7_14),
        ]

        # Tạo lưới các nút bấm
        for idx, (text, func) in enumerate(exercises):
            btn = tk.Button(frame_btn, text=text, command=func, width=30, height=1)
            btn.grid(row=idx // 2, column=idx % 2, padx=10, pady=5)

    # --- HÀM HỖ TRỢ TẠO CỬA SỔ CON ---
    def create_window(self, title):
        top = tk.Toplevel(self.root)
        top.title(title)
        top.geometry("400x350")
        return top

    # ============================================================
    # CHI TIẾT CÁC BÀI GIẢI
    # ============================================================

    def bai_7_1(self):
        # Bài 7.1: Viết chương trình GUI tạo một cửa sổ và đặt tiêu đề cho cửa sổ đó
        top = self.create_window("Bài 7.1")
        tk.Label(top, text="Đây là cửa sổ bài 7.1", font=FONT_LABEL).pack(pady=50)

    def bai_7_2(self):
        # Bài 7.2: Tạo cửa sổ, đặt tiêu đề và thêm một nhãn (Label)
        top = self.create_window("Bài 7.2")
        lbl = tk.Label(top, text="Xin chào! Đây là một Label.", font=FONT_LABEL)
        lbl.pack(pady=50)

    def bai_7_3(self):
        # Bài 7.3: Thay đổi kiểu phông chữ nhãn (đậm, kích thước)
        top = self.create_window("Bài 7.3")
        lbl_custom = tk.Label(top, text="Chữ Đậm Size 18", 
                              font=("Times New Roman", 18, "bold"), fg="red")
        lbl_custom.pack(pady=50)

    def bai_7_4(self):
        # Bài 7.4: Form điền thông tin sinh viên và nút gửi
        top = self.create_window("Bài 7.4 - Thông tin sinh viên")
        
        tk.Label(top, text="Tên sinh viên:").pack()
        entry_name = tk.Entry(top)
        entry_name.pack()
        
        tk.Label(top, text="ID sinh viên:").pack()
        entry_id = tk.Entry(top)
        entry_id.pack()
        
        tk.Label(top, text="Mật khẩu:").pack()
        entry_pass = tk.Entry(top, show="*")
        entry_pass.pack()

        def submit():
            info = f"Tên: {entry_name.get()}\nID: {entry_id.get()}\nPass: {entry_pass.get()}"
            messagebox.showinfo("Thông tin đã nhập", info)

        tk.Button(top, text="Gửi", command=submit).pack(pady=10)

    def bai_7_5(self):
        # Bài 7.5: Hiển thị file hình ảnh 'Otto.png'
        top = self.create_window("Bài 7.5 - Hiển thị ảnh")
        try:
            # Lưu ý: Bạn cần có file Otto.png cùng thư mục code để chạy được
            # Ở đây dùng PhotoImage của Tkinter (hỗ trợ PNG, GIF)
            img = tk.PhotoImage(file="Otto.png")
            lbl_img = tk.Label(top, image=img)
            lbl_img.image = img # Giữ tham chiếu để ảnh không bị mất
            lbl_img.pack()
        except Exception as e:
            tk.Label(top, text="Không tìm thấy file 'Otto.png'.\nVui lòng copy ảnh vào cùng thư mục code.", 
                     fg="red").pack(pady=50)

    def bai_7_6(self):
        # Bài 7.6: Nhập từ, nếu là 'Hello World' thì đảo ngược
        top = self.create_window("Bài 7.6 - Đảo chuỗi")
        
        tk.Label(top, text="Nhập một từ:").pack(pady=5)
        entry = tk.Entry(top)
        entry.pack(pady=5)
        lbl_result = tk.Label(top, text="")
        lbl_result.pack(pady=10)

        def dao_chuoi():
            text = entry.get()
            # Logic đảo chuỗi
            reversed_text = text[::-1]
            lbl_result.config(text=reversed_text)

        tk.Button(top, text="Validate (Đảo)", command=dao_chuoi).pack()

    def bai_7_7(self):
        # Bài 7.7: Tính tổng 1 + 2 + ... + N
        top = self.create_window("Bài 7.7 - Tính tổng")
        
        tk.Label(top, text="Nhập số nguyên N:").pack(pady=5)
        entry = tk.Entry(top)
        entry.pack()
        lbl_res = tk.Label(top, text="Tổng: ")
        lbl_res.pack(pady=10)

        def tinh_tong():
            try:
                n = int(entry.get())
                total = sum(range(1, n + 1))
                lbl_res.config(text=f"Tổng 1..{n} là: {total}")
            except ValueError:
                messagebox.showerror("Lỗi", "Vui lòng nhập số nguyên hợp lệ")

        tk.Button(top, text="Tính toán", command=tinh_tong).pack()

    def bai_7_8(self):
        # Bài 7.8: Liệt kê các ước số của N
        top = self.create_window("Bài 7.8 - Ước số")
        
        tk.Label(top, text="Nhập giá trị N:").pack(pady=5)
        entry = tk.Entry(top)
        entry.pack()
        
        lbl_res = tk.Label(top, text="Các ước số: ", wraplength=300)
        lbl_res.pack(pady=10)

        def tim_uoc():
            try:
                n = int(entry.get())
                uoc_so = [str(i) for i in range(1, n + 1) if n % i == 0]
                lbl_res.config(text="Các ước số: " + ", ".join(uoc_so))
            except ValueError:
                messagebox.showerror("Lỗi", "Nhập số nguyên dương")

        tk.Button(top, text="Validate", command=tim_uoc).pack()

    def bai_7_9(self):
        # Bài 7.9: Tính GCD (Ước chung lớn nhất) và LCM (Bội chung nhỏ nhất)
        top = self.create_window("Bài 7.9 - GCD & LCM")
        
        tk.Label(top, text="Nhập m:").pack(); e_m = tk.Entry(top); e_m.pack()
        tk.Label(top, text="Nhập n:").pack(); e_n = tk.Entry(top); e_n.pack()
        
        function_var = tk.StringVar(value="GCD")
        tk.Radiobutton(top, text="GCD (Ước chung max)", variable=function_var, value="GCD").pack()
        tk.Radiobutton(top, text="LCM (Bội chung min)", variable=function_var, value="LCM").pack()
        
        lbl_ketqua = tk.Label(top, text="Kết quả: ")
        lbl_ketqua.pack(pady=10)

        def tinh():
            try:
                m = int(e_m.get())
                n = int(e_n.get())
                gcd_val = math.gcd(m, n)
                
                if function_var.get() == "GCD":
                    res = gcd_val
                else:
                    res = abs(m * n) // gcd_val
                lbl_ketqua.config(text=f"{function_var.get()}({m},{n}) = {res}")
            except ValueError:
                messagebox.showerror("Lỗi", "Dữ liệu nhập sai")

        tk.Button(top, text="Tính", command=tinh).pack()

    def bai_7_10(self):
        # Bài 7.10: Kiểm tra tuổi (Thành niên hay chưa)
        top = self.create_window("Bài 7.10 - Kiểm tra tuổi")
        
        tk.Label(top, text="Tên của bạn:").pack(); e_name = tk.Entry(top); e_name.pack()
        tk.Label(top, text="Tuổi:").pack(); e_age = tk.Entry(top); e_age.pack()
        
        def check_age():
            try:
                name = e_name.get()
                age = int(e_age.get())
                if age >= 18:
                    msg = f"Xin chào {name}! Bạn đã trưởng thành."
                else:
                    msg = f"Xin chào {name}! Bạn còn nhỏ tuổi (vị thành niên)."
                messagebox.showinfo("Kết quả", msg)
            except ValueError:
                messagebox.showerror("Lỗi", "Tuổi phải là số")

        tk.Button(top, text="Kiểm tra", command=check_age).pack(pady=10)

    def bai_7_11(self):
        # Bài 7.11: Chuyển năm Dương lịch sang Âm lịch (Can Chi)
        top = self.create_window("Bài 7.11 - Đổi năm Âm lịch")
        
        tk.Label(top, text="Nhập năm Dương lịch:").pack(pady=5)
        e_year = tk.Entry(top)
        e_year.pack()
        
        lbl_am_lich = tk.Label(top, text="Năm Âm lịch: ", font=("Arial", 12, "bold"))
        lbl_am_lich.pack(pady=20)

        def doi_nam():
            try:
                nam_duong = int(e_year.get())
                # Mảng Can: số cuối của năm dương tương ứng với Can
                can_arr = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
                # Mảng Chi: chia dư cho 12 (0 là Thân)
                chi_arr = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
                
                can = can_arr[nam_duong % 10]
                chi = chi_arr[nam_duong % 12]
                
                lbl_am_lich.config(text=f"Năm Âm lịch: {can} {chi}")
            except ValueError:
                messagebox.showerror("Lỗi", "Vui lòng nhập năm hợp lệ")

        tk.Button(top, text="Chuyển đổi", command=doi_nam).pack()

    def bai_7_12(self):
        # Bài 7.12: Tính chỉ số BMI
        top = self.create_window("Bài 7.12 - Tính BMI")
        
        tk.Label(top, text="Cân nặng (kg):").pack(); e_w = tk.Entry(top); e_w.pack()
        tk.Label(top, text="Chiều cao (m):").pack(); e_h = tk.Entry(top); e_h.pack()
        lbl_bmi = tk.Label(top, text="BMI: ...")
        lbl_bmi.pack(pady=10)

        def calc_bmi():
            try:
                w = float(e_w.get())
                h = float(e_h.get())
                bmi = w / (h ** 2)
                
                status = ""
                if bmi < 18.5: status = "Gầy"
                elif bmi < 24.9: status = "Bình thường"
                elif bmi < 29.9: status = "Thừa cân"
                else: status = "Béo phì"
                
                lbl_bmi.config(text=f"BMI: {bmi:.2f}\nKết luận: {status}")
            except ValueError:
                messagebox.showerror("Lỗi", "Nhập số thực (ví dụ 1.75)")

        tk.Button(top, text="Tính BMI", command=calc_bmi).pack()

    def bai_7_13(self):
        # Bài 7.13: Xây dựng Menu hệ thống bài tập
        top = self.create_window("Bài 7.13 - Menu Demo")
        top.geometry("500x300")
        
        tk.Label(top, text="Hãy nhìn lên thanh Menu phía trên cửa sổ này", font=FONT_LABEL).pack(pady=50)
        
        # Tạo thanh menu
        menubar = Menu(top)
        top.config(menu=menubar)

        # Menu Hệ thống
        sys_menu = Menu(menubar, tearoff=0)
        menubar.add_cascade(label="Hệ thống bài tập Python nâng cao", menu=sys_menu)
        
        # Tạo Menu Chương 1 -> Bài 1.1 -> 1.1.1
        # Theo đề bài: Chương 1 -> 1.1 -> 1.1.1 gọi hộp thoại
        def show_demo_msg():
            messagebox.showinfo("Giải phương trình bậc 2", "Đây là hộp thoại demo chức năng 1.1.1")

        # Menu con cấp 2
        sub_menu_1_1 = Menu(sys_menu, tearoff=0)
        sub_menu_1_1.add_command(label="1.1.1 Giải phương trình bậc 2", command=show_demo_msg)
        
        # Menu con cấp 1
        sub_menu_1 = Menu(sys_menu, tearoff=0)
        sub_menu_1.add_cascade(label="1.1 Các bài tập cơ bản", menu=sub_menu_1_1)
        
        sys_menu.add_cascade(label="Chương 1", menu=sub_menu_1)
        sys_menu.add_separator()
        sys_menu.add_command(label="Thoát", command=top.destroy)

    def bai_7_14(self):
        # Bài 7.14: Tính cước Taxi
        top = self.create_window("Bài 7.14 - Tính tiền Taxi")
        
        tk.Label(top, text="Nhập số km đi được:").pack(); e_km = tk.Entry(top); e_km.pack()
        tk.Label(top, text="Thời gian chờ (phút):").pack(); e_time = tk.Entry(top); e_time.pack()
        
        lbl_tien = tk.Label(top, text="Thành tiền: 0 VND", font=("Arial", 14, "bold"), fg="red")
        lbl_tien.pack(pady=20)
        
        def tinh_tien():
            try:
                km = float(e_km.get())
                tg_cho = int(e_time.get())
                tien = 0
                
                # 1. Tính tiền di chuyển
                if km <= 0.8:
                    tien = 12000
                elif km <= 30:
                    # 12000 cho 0.8km đầu + phần còn lại * 15300
                    tien = 12000 + (km - 0.8) * 15300
                else:
                    # 12000 đầu + (30 - 0.8) * 15300 + (km - 30) * 12100
                    tien = 12000 + (29.2 * 15300) + (km - 30) * 12100
                
                # 2. Tính tiền chờ (miễn phí 5 phút đầu)
                if tg_cho > 5:
                    tien_cho = (tg_cho - 5) * 750
                    tien += tien_cho
                
                lbl_tien.config(text=f"Thành tiền: {int(tien):,} VND")
            except ValueError:
                messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ")

        tk.Button(top, text="Tính tiền", command=tinh_tien).pack()

# --- CHẠY CHƯƠNG TRÌNH ---
if __name__ == "__main__":
    root = tk.Tk()
    app = GiaiBaiTapChuong7(root)
    root.mainloop()
```
